Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    # Determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example Usage
likes = [200, 300, 250, 300]
analyze_post_engagement(likes)
```

Problem 3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example Usage
rainfall = [70, 75, 80, 63]
rainfall_checker(rainfall, 70)
```

Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_usernames(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)


# Example Usage
users = ["john", "alice", "bob", "john"]
detect_duplicate_usernames(users)
```

Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


# Example Usage
crop_prices = [1500, 2500, 3200, 1800]
filter_premium_crops(crop_prices)
```

Problem 7: System Monitoring – Application Health Checker

```python
def application_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)


# Example Usage
application_health_checker(7)
```

Problem 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for day in attendance_list:
        if day == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# Example Usage
attendance = [1, 1, 0, 1, 1]  # 1 = Present, 0 = Absent
attendance_eligibility(attendance)
```

Problem 10: Smart Electricity Bill Analyzer

```python
def electricity_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("Total Units:", units)
    print("Total Bill:", bill)
    print("Usage Level:", status)


# Example Usage
electricity_bill(250)
```
