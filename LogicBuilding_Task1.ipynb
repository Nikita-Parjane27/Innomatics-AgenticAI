```python
# Task 1: User Login Check

# Stored credentials
correct_username = "admin"
correct_password = "1234"

# Input from user
username = input("Enter username: ")
password = input("Enter password: ")

# Login validation
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# Task 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Analyze marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print result
print("Total Students Passed:", pass_count)
print("Total Students Failed:", fail_count)
```

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean names
for name in names:
    cleaned = name.strip().lower()  # remove spaces and convert to lowercase
    cleaned_names.append(cleaned)

print("Cleaned Names List:", cleaned_names)
```

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print("⚠️ Long Message Detected")
```

```python
# Task 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count errors
for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
