Problem 1: Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0

    # Process vehicle logs
    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT" and current_parked > 0:
            current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Check parking status
    if current_parked >= capacity:
        status = "Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)


# Example Usage
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
smart_parking_system(capacity, logs)
```

Problem 2: Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    # Base time calculation (5 minutes per km)
    eta = distance * 5

    # Traffic impact
    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 8

    # Weather impact
    if weather == "Rainy":
        eta += 10
    elif weather == "Storm":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")


# Example Usage
delivery_time_estimator(8, "High", "Rainy")
```

Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):
    booked_count = 0

    # Count booked seats
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", str(round(occupancy)) + "%")
    print("Show Status:", status)


# Example Usage
total_seats = 200
booked = [1] * 150
theatre_occupancy(total_seats, booked)
```

Problem 4: Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    # Classification
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", str(round(average_cpu)) + "%")
    print("Server Status:", status)


# Example Usage
cpu_data = [45, 60, 70, 85, 90]
server_load_classifier(cpu_data)
```

Problem 5: Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    # Define threshold for overuse (e.g., >8 hours)
    threshold = 8

    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", alert)


# Example Usage
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
classroom_resource_monitor(resources)
```

Problem 6: Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    confirmed = 0

    # Count confirmed registrations up to capacity
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1

    waitlisted = registrations - confirmed

    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted if waitlisted > 0 else 0)
    print("Registration Status:", status)


# Example Usage
event_registration_controller(100, 105)
```
